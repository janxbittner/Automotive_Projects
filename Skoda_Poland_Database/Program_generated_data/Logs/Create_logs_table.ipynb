Create dataframe with sales by month in 2020:

```python
sales_mth = pd.read_csv('Sales_by_month.txt',header = None)
del sales_mth[1]
sales_mth.columns = ['Sales']
sales_mth['pct'] = sales_mth.Sales / np.sum(sales_mth.Sales)
sales_mth
```

Function to create random day from given month:

```python
import datetime
import random
def random_date(number_of_month):
    start_date = datetime.date(2020, number_of_month+1, 1)
    if number_of_month == 11:
        end_date = datetime.date(2021, 1, 1)
    else:
        end_date = datetime.date(2020, number_of_month+2, 1)
        
    
    return np.random.choice(pd.date_range(start_date,end_date,freq='D').strftime("%Y-%m-%d %H:%M:%S").tolist())
```

Create dataframe with logs:

```python
# Create empty data frame
log_df = pd.DataFrame()

# Create dataframe with car model and annual sales 
sales_mth_model = pd.DataFrame(columns = ["Model",'Sales'])
sales_mth_model.Model = sales_df.car_model
sales_mth_model['Sales'] = sales_df.sale_in_2020

# Loop to append each row to log dataframe
for month_nb, sale_rating in enumerate (sales_mth['pct'].tolist()):
    sales_mth_model['Sales'] = (sales_df.sale_in_2020 * sale_rating).astype(int)
    for car_sales in sales_mth_model.values:
        nb_of_logs_of_model = car_sales[1]* stats_len
        for car in range(nb_of_logs_of_model):
            result_list = []
            temp = sales_df[sales_df['car_model'] == car_sales[0]]
            result_list.append(np.random.choice(temp.car_model_id.item(),1).tolist()[0]) # ID OF CAR
            result_list.append(np.random.choice(list(range(len(statuses))))) # ID OF STATUS
            result_list.append(np.random.choice(list(range(dealer_len)))) # ID OF DEALERSHIP
            result_list.append(random_date(month_nb)) # DATE OF LOG
            log_df = log_df.append(pd.DataFrame(result_list).T) # Append result list to log dataframe
log_df
```

Save database to csv:

```python
log_df.to_csv('final_logs_database.csv',sep =',',header = False)
```
