Import libraries

```python
import pandas as pd
import numpy as np
```

Import database

```python
raw = pd.read_csv('Dealerships.csv',encoding = "ISO-8859-1",delimiter=';')
raw
```

Import street names form .txt

```python
street = pd.read_csv('streets.txt',delimiter=',',header=None,names=['StreetNames','1'])
del street['1']
```

Generate random address:

```python
def generate_address():
    return str(np.random.choice(street.StreetNames)) + " " + str(*np.random.randint(0,1000,(1)))
```

Add random addresses to database:

```python
raw.address = raw.address.apply(lambda x :generate_address())
raw
```

Import dilership names form .txt

```python
names = pd.read_csv('Dealership_names.txt',delimiter='/n',header=None,names=['Names','1'],engine='python')
del names['1']
names.head(3)
```

Generate random dealership Names:

```python
def generate_name():
    return str(np.random.choice(names.Names))
```

Add random names to database:

```python
raw.Name = raw.Name.apply(lambda x :generate_name())
raw
```

Generate and add random phone number:

```python
raw['phone number'] = raw['phone number'].apply(lambda x :'+48 '+ str(np.random.randint(120000000,950000000, dtype=np.int64)))
raw
```

Generate and add email to database:

```python
raw['email'] = raw['Name'] + '_'+ raw['City'] + '@Skoda.pl'
raw
```

Remove spaces from email:

```python
raw['email'] = raw['email'].apply(lambda x :x.replace(' ','').lower())
raw
```

Save database to csv:

```python
raw.to_csv('finall_dealership.csv',sep =',',header = False)
```
